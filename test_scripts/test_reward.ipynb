```python
import numpy as np

import matplotlib.pyplot as plt

# Define constants
Cr = 1

# Define the range of Dt
Dt = np.linspace(-10, 10, 500)  # Avoid division by zero by starting from 0.1
Dt_l = np.linspace(-10, 10, 500) + 10 * np.linspace(-1, 1, 500)
# Define the function
y_Dt = Cr * (Dt_l - Dt) ** 3

x = np.linspace(-10, 10, 500)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y_Dt, label=r'$Cr \cdot (Dt_l - Dt)^3$')
plt.plot(Dt, Dt, label=r'Dt')
plt.plot(Dt, Dt_l, label=r'Dt_l')
plt.xlabel('Dt')
plt.ylabel('y')
plt.title('Plot of the function')
plt.legend()
plt.grid()
plt.show()
```
